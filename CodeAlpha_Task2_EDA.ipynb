```python
# Import required libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('IPL2025Batters.csv')

# Replace '-' with NaN (missing values)
df.replace('-', np.nan, inplace=True)
```

```python
numeric_cols = ['Runs', 'Matches', 'Inn', 'No', 'AVG', 'BF', 'SR',
                '100s', '50s', '4s', '6s']

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df[numeric_cols] = df[numeric_cols].fillna(0)

print(df.shape)          # Number of rows and columns
print(df.dtypes)         # Data type of each column
print(df.head())         # First 5 rows
print(df.describe())     # Statistical summary
print(df.isnull().sum()) # Missing values in each column
```

```python
# Clean HS column (remove * from scores like 98*)
df['HS_clean'] = df['HS'].str.replace('*', '', regex=False).astype(int)

# Create new columns
df['Boundaries'] = df['4s'] + df['6s']
df['Balls_per_Innings'] = (df['BF'] / df['Inn']).round(2)
df['Dot_Ratio'] = ((df['BF'] - df['Runs']) / df['BF']).round(3)

# Display first 5 rows
print(df[['Player Name','HS','HS_clean','Boundaries','Balls_per_Innings','Dot_Ratio']].head())
```

```python
# Top 10 Run Scorers
print("Top 10 Run Scorers")
print(df[['Player Name', 'Team', 'Runs']].sort_values(by='Runs', ascending=False).head(10))

print("\n")

# Top 10 Strike Rate
print("Top 10 Strike Rate")
print(df[['Player Name', 'Team', 'SR']].sort_values(by='SR', ascending=False).head(10))

print("\n")

# Top 10 Average
print("Top 10 Batting Average")
print(df[['Player Name', 'Team', 'AVG']].sort_values(by='AVG', ascending=False).head(10))
```

```python
# Team-wise total runs
team_runs = df.groupby('Team')['Runs'].sum().sort_values(ascending=False)

print("Team-wise Total Runs")
print(team_runs)
```

```python
# Correlation between numeric columns
correlation = df[numeric_cols].corr()

print(correlation)
```

```python
print("----- DATASET INSIGHTS -----")

print(f"Total Players : {len(df)}")
print(f"Highest Run Scorer : {df.loc[df['Runs'].idxmax(),'Player Name']}")
print(f"Highest Runs : {df['Runs'].max()}")

print(f"Highest Strike Rate : {df.loc[df['SR'].idxmax(),'Player Name']}")
print(f"Strike Rate : {df['SR'].max()}")

print(f"Highest Average : {df.loc[df['AVG'].idxmax(),'Player Name']}")
print(f"Average : {df['AVG'].max()}")

print(f"Total Teams : {df['Team'].nunique()}")
```
